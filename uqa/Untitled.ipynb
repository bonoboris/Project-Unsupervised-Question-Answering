```python
import json

# Load a datastore
with open('wiki_00.json', 'r') as f:
    datastore = json.load(f)
```

```python
import time
# Get the first article
article = datastore[0]

for article in datastore:
    contexts = article['contexts']
    for context in contexts:
        text = context['text']
        entities = context['entities']
        tree_text = list(parser.parse_text(text))
        time.sleep(5)
        for tree in tree_text:
            print(tree)
```
